### 3. Introducing [Linear Regression](https://scikit-learn.org/stable/modules/linear_model.html) with Scikit Learn

```python
# Import sklearn
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Define model

regr = linear_model.LinearRegression()
```

```python
# Define label and features (adjustable)

features = ['BEDS', 'BATHS', 'SQFT', 'AGE', 'LOTSIZE', 'GARAGE']
label = ['SOLDPRICE']
```

```python
# Filter out empties

filter_data = data.copy()

# Additional
filter_data = filter_data[filter_data['SOLDPRICE'] < 2000000]


# Apply to all
for f in features + label:
    # Out empties - (v. imputation)
    filter_data = filter_data[filter_data[f].notna()]
    
    # Interaction terms
    
    # Standardize - don't need
    # filter_data[f] = (filter_data[f] - filter_data[f].mean()) / filter_data[f].std()
    
    # v. Normalize - don't need either (but can add in regression constructor)
    # filter_data[f] = (filter_data[f] - filter_data[f].min()) / (filter_data[f].max() - filter_data[f].min())
    
filter_data = filter_data.reset_index()

print(filter_data[label + features].head())
```

```python
# Extract features and label

x = filter_data[features]
y = filter_data[label]
```

```python
# !YOUR TURN!

# Take a look at the shapes for 'labels' and 'features'
"labels", y.shape, "features", x.shape
```

```python
# Set up testing and training (adjustable)
split = int(len(filter_data) * .7)

x_train = x[:split]
x_test = x[split:]
y_train = y[:split]
y_test = y[split:]

# Check shapes

print ("Train shape")
print (x_train.shape, y_train.shape)
print ("\nTest shape")
print (x_test.shape, y_test.shape)
```

```python
# Fit training sets

regr.fit(x_train, y_train)
```

```python
# Make predictions using testing set

y_pred = regr.predict(x_test)

# Explained variance score: 1 is perfect prediction

print(f'Variance score: {round(r2_score(y_test, y_pred), 2)}')
```

### 4. Plot, aka "Visualize"

```python
# Import
import matplotlib.pyplot as plt
```

```python

# Plot outputs
plt.figure(figsize=(20, 5))
plt.scatter(y_test, y_pred, color='black')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'b--', lw=4)

plt.ylabel('Predicted')
plt.xlabel('Measured')

plt.xticks(())
plt.yticks(())

plt.show()
```

```python
# Residual plot

plt.figure(figsize=(10, 7))

resid = y_test - regr.predict(x_test)

plt.axhline(y=0, linestyle='-', linewidth=2, color="r")
plt.scatter(x=y_pred, y=resid, alpha=0.5, s=3)

plt.title("Residual plot")
plt.ylabel(r"Y-$\hat Y$")
plt.xlabel(r'$\hat Y$')
plt.tight_layout()
plt.show()
```

### 7. Export

```python
# Download model as pkl file
import joblib
joblib.dump(regr, 'regr.pkl')
```

```python
print(r2_score(y_test, regr.predict(x_test)))
print(mean_squared_error(y_test, regr.predict(x_test)))
```

### Use different features and/or different train/test split to improve your model

```python
# YOUR TURN
# Use sklearn train|test|split
# Ref. https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html
from sklearn.model_selection import train_test_split


# from above dataframe, select different features and different filter
new_df = data[['SOLDPRICE', 'LISTPRICE', 'SQFT', 'BEDS', 'BATHS', 'AGE', 'LOTSIZE']].dropna()


# # Init x, y test,train @ different size than above (sklearn) (e.g. 33%)
X_train, X_test, y_train, y_test = train_test_split(new_df[['LISTPRICE', 'SQFT', 'BEDS', 'BATHS', 'AGE', 'LOTSIZE']], new_df['SOLDPRICE'], test_size=0.33, random_state=42)



# Determine linear_model 2 (sklearn - above) (do not use the same name for the model)

# This linear regression is likely better because I specifically looked for attributes that I thought would be more relevent 
# to the selling price. This was helpful because that way the moel would not get "distracted" with unimportant traits.
new_regression_model = linear_model.LinearRegression()



# Train linear_model (sklearn)
new_regression_model.fit(X_train, y_train)


# Score Method 1: (sklearn)
# Ref. https://stackoverflow.com/questions/45529907/difference-between-model-score-vs-r2-score
#  https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html
new_score_using_r2 = r2_score(y_test, new_regression_model.predict(X_test))

# Score Method 2 (sklearn)
new_score_using_mse = mean_squared_error(y_test, new_regression_model.predict(X_test))


# Plot outputs (matplotlib - above)
plt.figure(figsize=(20, 5))
plt.scatter(y_test, new_regression_model.predict(X_test), color='black')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'b--', lw=4)

# Download model 2 as pkl file
joblib.dump(new_regression_model, 'new_regression_model.pkl')
```

```python
new_score_using_r2, new_score_using_mse
```

The r2 score of the new model is higher and the mse is lower.
